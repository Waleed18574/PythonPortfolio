## Introduction
Concrete is the most important material in civil engineering and it is  of five major elements in various proportions: cement, water, coarse aggregates, fine aggregates (i.e. sand), and air. Concrete admixtures are natural or manufactured chemicals or additives added during concrete mixing to enhance specific properties of the fresh or hardened concrete, such as workability, durability, or early and final strength. The concrete compressive strength is a highly nonlinear function of age and ingredients. The have been several research to develop the optimum mixure of Concrete. 


## Problem Statement
The Concrete research usually uses Concrete Compressive Strength Test in lab to find out the resultant strengths of different combination of concrete ingredients. This can be costly and time consuming. In this project, I will predict the compressive strength of concrete mixure measured in MPa using the historical records of previous testing and regression analysis methods.

### Features Non-normality

Some beginners think (erroneously) that the normal distribution assumption of linear regression applies to their data. They might plot their response variable as a histogram and examine whether it differs from a normal distribution. Others assume that the explanatory variable must be normally-distributed. Neither is required. The normality assumption relates to the distributions of the residuals. This is assumed to be normally distributed, and the regression line is fitted to the data such that the mean of the residuals is zero.
\
[Source](https://mickteaching.wordpress.com/2016/04/19/data-need-to-be-normally-distributed-and-other-myths-of-linear-regression/)

### Removing Outliers From  Target

It can be seen that the target contains some outliers. Removing those outliers can imporove the scoring of the model. There are two methods for detecting and removing outliers. For normally distributed data I can use (mean - 3*standard_deviation) and (mean + 3*standard_deviation). For non normally distibuted data, I can use the interquartile method.Since the data is not normally distributed, I will use the interquartile method

```python
q1 = df.compressive_strength_csMPa.quantile(0.25)
q3 = df.compressive_strength_csMPa.quantile(0.75)
iqr = q3-q1
upper_lim = q3 + 1.5*iqr
lower_lim = q1 - 1.5*iqr

df = df[(df.compressive_strength_csMPa >= lower_lim) & (df.compressive_strength_csMPa <= upper_lim)]
```

```python
col = 'compressive_strength_csMPa'

plt.figure(figsize=(10,4))
plt.suptitle(f'{col} Distribution')

plt.subplot(1,2,1)
sns.histplot(data=df, x = col, bins=30)
plt.title('Histogram')

plt.subplot(1,2,2)
sns.boxplot(data=df, x = col)
plt.title('Boxplot')

plt.show()
```

## Features Selection (based on statiscal testing)

```python
features = [col for col in df.columns if col != 'compressive_strength_csMPa']
y = df['compressive_strength_csMPa']
```

```python
significant_features = []
for col in features:
    corr, p_value = pearsonr(df[col],df['compressive_strength_csMPa'])
    print(f'{col} vs. compressive_strength_csMPa:\n')
    print('Correlation Coefficient:',corr)
    print('p_value:',p_value)

    if p_value < 0.05:
        significant_features.append(col)
        if corr > 0:
            print(f'There is a significant evidence of a positive relationship between {col} and Concrete Compressive Strength')
        else:
            print(f'There is a significant evidence of a negative relationship between {col} and Concrete Compressive Strength')
    else:
        print(f'There is no significant relationship between {col} and Concrete Compressive Strength')

    plt.figure(figsize = (6,4), facecolor='lightblue')
    plt.title(f'{col.capitalize()} vs. Compressive Strength of Concrete')
    sns.regplot(x=col, y='compressive_strength_csMPa',data=df)
    plt.show()
```

## Validating the Model

### Checking for the Normality of the Residual

```python
# using qqplot
sm.qqplot(lin_reg_model.resid, line='s')
pylab.title('QQPlot of the Residual')
pylab.show()
```

```python
plt.figure(figsize=(18,4))

plt.subplot(1,2,1)
plt.title('Histogram of the Residual')
sns.histplot(lin_reg_model.resid)

plt.subplot(1,2,2)
plt.title('KDE of the Residual')
sns.kdeplot(lin_reg_model.resid)

plt.show()
```

### Checking for Heteroscedasticity of the Residual

__H0:__ The variances of the errors are equal

__H1:__ The variances of the errors are not equal

alpha = 0.05

```python
# using Breusch-Pegan Test
_, p_value,_,f_pvalue = diag.het_white(lin_reg_model.resid, lin_reg_model.model.exog)
print('P-vlaue:',p_value)
print('F-Statisticse:',f_pvalue)
```

### Checking for Autocorrelation

Using Ljung-Box

H0: There is no autocorrelation between the errors

H1: There is an autocorrelation betwene the erros

alpha = 0.05

```python
# calculating the lags
lags = min(10, len(x_train)//5) 

# conduct the test
test_result = diag.acorr_ljungbox(lin_reg_model.resid, lags = lags)

# catch the p_value
stats, p_vals = test_result.iloc[:,0],test_result.iloc[:,1]

#print the result of the test
if min(p_vals) > 0.05:
    print(f'Minimum P-value : {min(p_vals):0.4}')
    print('There is significant statistical evidence there is no autocorrelation between the erros')
else:
    print(f'Minimum P-value : {min(p_vals):0.4}')
    print('There is significant statistical evidence there is an autocorrelation between the erros')
    
```

```python
sm.graphics.tsa.plot_acf(lin_reg_model.resid, lags = lags)
plt.title('Autocorrelation fo the Residual')
plt.show()
```

## Other Models

Next, I will loop through different models and see the models with the largest r2 value

```python
# initiate the mdoels
ridge=Ridge()
svr=LinearSVR()
knn=KNeighborsRegressor()
dt=DecisionTreeRegressor()
xg_re=xgb.XGBRegressor(objective='reg:linear',n_estimators=10)
ad=AdaBoostRegressor(base_estimator=dt,n_estimators=400)
gbr=GradientBoostingRegressor(n_estimators=300,max_depth=1)
rfr=RandomForestRegressor(n_estimators=400)


# make a dictionary of the models names and the models
models={'Ridge Regression':ridge,
        'Support Vector Machine':svr,
        'K Nearest Neighbours':knn,
        'Decision Tree':dt,
        'Extreme Gradient Boosting':xg_re,
        'Ada Boost':ad,
        'Gradient Boosting':gbr,
        'Random Forest':rfr}
```

```python
mse = {} 
r2 = {}

# loop through the models and register mse and r2
for model_name, model in models.items():
    model.fit(x_train, y_train)
    y_pred1 = model.predict(x_test)
    mse[model_name] = mean_squared_error(y_test,y_pred1)
    r2[model_name] = r2_score(y_test,y_pred1)
    
```

```python
# convert mse and r2 dictionaries to dataframes
mse = pd.Series(mse).to_frame().rename(columns={0:'mean squared error'}) 
r2 = pd.Series(r2).to_frame().rename(columns={0:'r2 score'})

# concatinate the two dataframes into mse_r2 dataframe
mse_r2 = pd.concat([mse,r2], axis = 1)
mse_r2.sort_values('r2 score', ascending = False, inplace = True)
mse_r2 
```

```python
# select the best model based on the r2_score
best_model = mse_r2.index[0]
best_model_r2_score  = mse_r2['r2 score'][0]
print(f'Best Model is {best_model} with r2_score of {best_model_r2_score:.2}')

```

```python
best_model = models[best_model]

# redo the best model.
best_model.fit(x_train,y_train)
y_pred1 = best_model.predict(x_test)
```
